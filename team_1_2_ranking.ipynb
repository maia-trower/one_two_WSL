```python
import pickle
import pandas as pd
import numpy as np
from statsbombpy import sb
import seaborn as sns
import matplotlib.pyplot as plt
pd.set_option("display.max_columns", None)
```

```python
with open("data/WSL_20_21_all_one_twos.pickle", "rb") as f:
    one_twos = pickle.load(f)
```

```python
# count the number of one-twos per side

# get final standing of league
teams_standing = ["Chelsea FCW", "Manchester City WFC", "Arsenal WFC", "Manchester United", "Everton LFC",
                  "Brighton & Hove Albion WFC", "Reading WFC", "Tottenham Hotspur Women",
                  "West Ham United LFC", "Aston Villa", "Birmingham City WFC", "Bristol City WFC"]
# then put it all in a single df to count easier
one_twos_aggregated = pd.DataFrame([])
for key in one_twos.keys():
    one_twos_aggregated = pd.concat([one_twos_aggregated, one_twos[key]], axis=0)
```

```python
team_counts = one_twos_aggregated.team.value_counts()
team_counts = team_counts.reset_index()

team_counts["count"] = np.divide(team_counts["count"], 2)
```

```python
sns.barplot(x="team", y="count", data=team_counts)
plt.title("Number of One-Two Passes Per Team (WSL 20/21)")
plt.ylabel("Count")
plt.xlabel("Team")
plt.xticks(np.arange(len(team_counts)), teams_standing, rotation=90)
plt.show()
```
